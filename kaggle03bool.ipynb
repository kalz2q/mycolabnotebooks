````
# Define a function which takes a numerical argument and returns -1 if it's negative, 1 if it's positive and 0 otherwise
def sign (x) :
    if x < 0 :
        return -1
    elif x > 0 :
        return 1
    else :
        return 0

print (sign(-1))
print (sign(1))
print (sign(0))
````
