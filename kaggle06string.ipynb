### å•é¡Œ 2. ãƒ•ã‚£ãƒ«ã‚¿ãƒªãƒ³ã‚°

ã¨ã‚ã‚‹ç ”ç©¶è€…ãŒæ•°åƒã®ãƒ‹ãƒ¥ãƒ¼ã‚¹è¨˜äº‹ã‚’é›†ã‚ãŸã€‚ç‰¹å®šã®å˜èªžã§è¨˜äº‹ã‚’ãƒ•ã‚£ãƒ«ã‚¿ãƒ¼ã«ã‹ã‘ãŸã„ã¨è€ƒãˆã¦ã„ã‚‹ã€‚ãã®ãŸã‚ã®é–¢æ•°ã‚’ä½œã‚‹ã€‚

é–¢æ•°ã¯æ¬¡ã®åŸºæº–ã‚’æº€ãŸã™ã‚‚ã®ã¨ã™ã‚‹ã€‚

- ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰ã¯å˜ç‹¬ã§ç¾ã‚Œã‚‹ã‚‚ã®ã ã‘ã‚’ã¨ã‚‹ã€‚ã¤ã¾ã‚Š"closed"ã ã£ãŸã‚‰ã€"enclosed"ã¯å«ã¾ã‚Œãªã„ã€‚
- å¤§æ–‡å­—å°æ–‡å­—ã¯åŒºåˆ¥ã—ãªã„ã€‚ã¤ã¾ã‚Šã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰ãŒ"closed"ãªã‚‰ã°"Closed the case."ã¯å«ã¾ã‚Œã‚‹ã€‚
- ãƒ”ãƒªã‚ªãƒ‰ã¯ãƒžãƒƒãƒã«å½±éŸ¿ã—ãªã„ã€‚"It is closed."ã¯å«ã¾ã‚Œã‚‹ã€‚ä»–ã®å¥èª­ç‚¹ã¯ãªã„ã‚‚ã®ã¨ã—ã¦ã‚ˆã„ã€‚

```
def word_search(doc_list, keyword):
    """
    Takes a list of documents (each document is a string) and a keyword. 
    Returns list of the index values into the original list for all documents 
    containing the keyword.

    Example:
    doc_list = ["The Learn Python Challenge Casino.", "They bought a car", "Casinoville"]
    >>> word_search(doc_list, 'casino')
    >>> [0]
    """
    match = [doc.lower().strip(".").split().count(keyword.lower()) for doc in doc_list]
    dictOfMatch = { i : match[i] for i in range(0, len(match))}
    return  [key for key in dictOfMatch if dictOfMatch[key] > 0]

doc_list = ["The Learn Python Challenge Casino.", "They bought a car", "Casinoville"]
word_search(doc_list, 'casino')
```

## å•é¡Œ 3 è¤‡æ•°ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰æ¤œç´¢

æ¬¡ã«ã€ç ”ç©¶è€…ã¯è¤‡æ•°ã®ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰ã§æ¤œç´¢ã—ãŸã„ã€ã¨æ€ã£ãŸã€‚

ã„ã¾ä½œã£ãŸ`word_search`ã‚’åˆ©ç”¨ã—ã¦æ¬¡ã®é–¢æ•°ã‚’ä½œã‚‹ã“ã¨ã€‚


```
def multi_word_search(doc_list, keywords):
    """
    Takes list of documents (each document is a string) and a list of keywords.  
    Returns a dictionary where each key is a keyword, and the value is a list of indices
    (from doc_list) of the documents containing that keyword

    >>> doc_list = ["The Learn Python Challenge Casino.", "They bought a car and a casino", "Casinoville"]
    >>> keywords = ['casino', 'they']
    >>> multi_word_search(doc_list, keywords)
    {'casino': [0, 1], 'they': [1]}
    """
    dictOfMatch = { keyword : word_search(doc_list, keyword) for keyword in keywords }
    
    return dictOfMatch

doc_list = ["The Learn Python Challenge Casino.", "They bought a car and a casino", "Casinoville"]
keywords = ['casino', 'they']

multi_word_search(doc_list, keywords)
```
